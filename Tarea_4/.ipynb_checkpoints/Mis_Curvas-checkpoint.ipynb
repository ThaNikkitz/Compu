```python
from matplotlib import pyplot as plt
%matplotlib inline
import numpy
```

```python
a = numpy.loadtxt(fname = 'data1.txt')
```

```python
b = []
c = []
for i in a:
    b.append(i[0])
    c.append(i[2])

plt.figure(figsize = (10,10))
plt.axhline(0, color = 'b', linestyle = ':')
for j in range(7):
    plt.plot(b[j*161:(j+1)*161], c[j*161:(j+1)*161])
```
