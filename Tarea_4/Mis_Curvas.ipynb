```python
from matplotlib import pyplot as plt
%matplotlib inline
import numpy
beta = [0.33, 0.5, 0.67, 0.91, 1.0, 1.33, 2.0]
```

```python
a1 = numpy.loadtxt(fname = 'data1.txt')
```

```python
b1 = []
c1 = []
for i in a1:
    b1.append(i[1])
    c1.append(i[2])

plt.figure(figsize = (10,8))
plt.title(r'Curvas de Amplitud en función de $\hat{\tau}$, con $\alpha$ = 50')
plt.ylabel('$A$', size = 20)
plt.xlabel(r'$\hat{\tau}$', size = 20)
plt.legend()
plt.axhline(0, color = 'k', linestyle = '-')
for j in range(7):
    plt.plot(b1[j*202:(j+1)*202], c1[j*202:(j+1)*202], linestyle = '-', label = r'$\beta$ = %02.2f' % beta[j])
    plt.legend()
plt.grid()
```
