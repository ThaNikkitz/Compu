```python
import numpy
from matplotlib import pyplot
%matplotlib inline 

a = numpy.linspace(0,212.8,110)
results = []
resultados = []
with open('resultados.txt') as inputfile:
    for line in inputfile:
        results.append(line.strip().split(','))
        resultados.append(line.strip())
for i in range(len(resultados)):
    resultados[i] = -float(resultados[i])
```

```python
pyplot.xlim(0, 212.8)
pyplot.ylim(-73, 73)
pyplot.title('Perfil del Gancho')
pyplot.xlabel('X $[mm]$')
pyplot.ylabel('Y $[mm]$')
pyplot.plot(a,results, color = 'k')
pyplot.plot(a,resultados, color = 'k');
```
